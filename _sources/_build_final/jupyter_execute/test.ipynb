```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Tạo dữ liệu giả lập
days = ['Thứ Hai', 'Thứ Ba', 'Thứ Tư', 'Thứ Năm', 'Thứ Sáu', 'Thứ Bảy', 'Chủ Nhật']
hours = range(24)

np.random.seed(0)
data = np.random.randint(50, 500, size=(7, 24))

# Tạo DataFrame
df = pd.DataFrame(data, index=days, columns=hours)

# Thiết lập kích thước biểu đồ
plt.figure(figsize=(12, 6))

# Vẽ Heat Map
sns.heatmap(df, cmap='YlOrRd', linewidths=.5, annot=True, fmt="d")

# Thiết lập tiêu đề và nhãn
# plt.title('Biểu đồ Heat Map lưu lượng truy cập website theo giờ và ngày trong tuần', fontsize=16)
# plt.xlabel('Giờ trong ngày', fontsize=12)
# plt.ylabel('Ngày trong tuần', fontsize=12)

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
labels = ['Apple', 'Samsung', 'Xiaomi', 'LG', 'Oppo', 'Khác']
sizes = [30, 25, 15, 10, 8, 12]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# Tạo biểu đồ Pie Chart
plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Vẽ hình tròn
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Thiết lập định dạng
# plt.title('Thị phần các hãng điện thoại di động năm 2023', fontsize=16)
plt.axis('equal')  # Đảm bảo biểu đồ là hình tròn

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify

# Dữ liệu
labels = ['Phần mềm\n500', 'Phần cứng\n400', 'Dịch vụ đám mây\n300',
          'Siêu thị\n300', 'Cửa hàng tiện lợi\n200', 'Thương mại điện tử\n300',
          'Dầu khí\n400', 'Năng lượng tái tạo\n200']

sizes = [500, 400, 300, 300, 200, 300, 400, 200]
colors = ['#8dd3c7','#ffffb3','#bebada','#fb8072','#80b1d3','#fdb462','#b3de69','#fccde5']

# Tính toán vị trí cho các ô
norm_sizes = [float(value) / sum(sizes) for value in sizes]
squarify_sizes = [size * 100 for size in norm_sizes]

# Vẽ biểu đồ
plt.figure(figsize=(12,6))
squarify.plot(sizes=squarify_sizes, label=labels, color=colors, alpha=.8)

# Thiết lập tiêu đề và bỏ khung
plt.title('Biểu đồ Tree Map cơ cấu doanh thu của tập đoàn năm 2023', fontsize=16)
plt.axis('off')

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
quy = ['Quý 1', 'Quý 2', 'Quý 3', 'Quý 4']
san_pham_A = [50, 60, 70, 80]
san_pham_B = [30, 35, 40, 45]
san_pham_C = [20, 25, 30, 35]

# Thiết lập vị trí cho các cột trên trục x
index = np.arange(len(quy))
width = 0.6  # Độ rộng của mỗi cột

# Vẽ biểu đồ Stacked Bar Chart
plt.figure(figsize=(10, 6))

plt.bar(index, san_pham_A, width, label='Sản phẩm A', color='#ff9999')
plt.bar(index, san_pham_B, width, bottom=san_pham_A, label='Sản phẩm B', color='#66b3ff')
bottom_C = np.array(san_pham_A) + np.array(san_pham_B)
plt.bar(index, san_pham_C, width, bottom=bottom_C, label='Sản phẩm C', color='#99ff99')

# Thiết lập tiêu đề và nhãn trục
# plt.title('Doanh số bán hàng theo quý và sản phẩm năm 2023', fontsize=16)
# plt.xlabel('Quý', fontsize=12)
# plt.ylabel('Doanh số (ngàn USD)', fontsize=12)
# plt.xticks(index, quy)

# Thêm chú thích
plt.legend()

# Hiển thị giá trị trên đỉnh mỗi cột
for i in range(len(quy)):
    total = san_pham_A[i] + san_pham_B[i] + san_pham_C[i]
    plt.text(x=index[i], y=total + 2, s=str(total), ha='center', va='bottom', fontsize=10)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu giả lập cho điểm thi
np.random.seed(0)
diem_thi = np.random.normal(70, 15, 200)  # Điểm trung bình 70, độ lệch chuẩn 15
diem_thi = np.clip(diem_thi, 0, 100)  # Giới hạn điểm trong khoảng từ 0 đến 100

# Vẽ biểu đồ Histogram
plt.figure(figsize=(10, 6))
plt.hist(diem_thi, bins=10, edgecolor='black', color='#66b3ff')

# Thiết lập tiêu đề và nhãn trục
plt.title('Phân bố điểm thi của sinh viên', fontsize=16)
plt.xlabel('Điểm số', fontsize=12)
plt.ylabel('Số lượng sinh viên', fontsize=12)

# Hiển thị biểu đồ
plt.show()
```
