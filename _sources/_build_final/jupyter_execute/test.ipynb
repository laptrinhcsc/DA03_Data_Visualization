```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# To d liu gi lp
days = ['Th Hai', 'Th Ba', 'Th T', 'Th Nm', 'Th Su', 'Th By', 'Ch Nht']
hours = range(24)

np.random.seed(0)
data = np.random.randint(50, 500, size=(7, 24))

# To DataFrame
df = pd.DataFrame(data, index=days, columns=hours)

# Thit lp kch thc biu 
plt.figure(figsize=(12, 6))

# V Heat Map
sns.heatmap(df, cmap='YlOrRd', linewidths=.5, annot=True, fmt="d")

# Thit lp tiu  v nhn
# plt.title('Biu  Heat Map lu lng truy cp website theo gi v ngy trong tun', fontsize=16)
# plt.xlabel('Gi trong ngy', fontsize=12)
# plt.ylabel('Ngy trong tun', fontsize=12)

# Hin th biu 
plt.show()
```

```python
import matplotlib.pyplot as plt

# D liu
labels = ['Apple', 'Samsung', 'Xiaomi', 'LG', 'Oppo', 'Khc']
sizes = [30, 25, 15, 10, 8, 12]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# To biu  Pie Chart
plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# V hnh trn
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Thit lp nh dng
# plt.title('Th phn cc hng in thoi di ng nm 2023', fontsize=16)
plt.axis('equal')  # m bo biu  l hnh trn

# Hin th biu 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify

# D liu
labels = ['Phn mm\n500', 'Phn cng\n400', 'Dch v m my\n300',
          'Siu th\n300', 'Ca hng tin li\n200', 'Thng mi in t\n300',
          'Du kh\n400', 'Nng lng ti to\n200']

sizes = [500, 400, 300, 300, 200, 300, 400, 200]
colors = ['#8dd3c7','#ffffb3','#bebada','#fb8072','#80b1d3','#fdb462','#b3de69','#fccde5']

# Tnh ton v tr cho cc 
norm_sizes = [float(value) / sum(sizes) for value in sizes]
squarify_sizes = [size * 100 for size in norm_sizes]

# V biu 
plt.figure(figsize=(12,6))
squarify.plot(sizes=squarify_sizes, label=labels, color=colors, alpha=.8)

# Thit lp tiu  v b khung
plt.title('Biu  Tree Map c cu doanh thu ca tp on nm 2023', fontsize=16)
plt.axis('off')

# Hin th biu 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D liu
quy = ['Qu 1', 'Qu 2', 'Qu 3', 'Qu 4']
san_pham_A = [50, 60, 70, 80]
san_pham_B = [30, 35, 40, 45]
san_pham_C = [20, 25, 30, 35]

# Thit lp v tr cho cc ct trn trc x
index = np.arange(len(quy))
width = 0.6  #  rng ca mi ct

# V biu  Stacked Bar Chart
plt.figure(figsize=(10, 6))

plt.bar(index, san_pham_A, width, label='Sn phm A', color='#ff9999')
plt.bar(index, san_pham_B, width, bottom=san_pham_A, label='Sn phm B', color='#66b3ff')
bottom_C = np.array(san_pham_A) + np.array(san_pham_B)
plt.bar(index, san_pham_C, width, bottom=bottom_C, label='Sn phm C', color='#99ff99')

# Thit lp tiu  v nhn trc
# plt.title('Doanh s bn hng theo qu v sn phm nm 2023', fontsize=16)
# plt.xlabel('Qu', fontsize=12)
# plt.ylabel('Doanh s (ngn USD)', fontsize=12)
# plt.xticks(index, quy)

# Thm ch thch
plt.legend()

# Hin th gi tr trn nh mi ct
for i in range(len(quy)):
    total = san_pham_A[i] + san_pham_B[i] + san_pham_C[i]
    plt.text(x=index[i], y=total + 2, s=str(total), ha='center', va='bottom', fontsize=10)

# Hin th biu 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# To d liu gi lp cho im thi
np.random.seed(0)
diem_thi = np.random.normal(70, 15, 200)  # im trung bnh 70,  lch chun 15
diem_thi = np.clip(diem_thi, 0, 100)  # Gii hn im trong khong t 0 n 100

# V biu  Histogram
plt.figure(figsize=(10, 6))
plt.hist(diem_thi, bins=10, edgecolor='black', color='#66b3ff')

# Thit lp tiu  v nhn trc
plt.title('Phn b im thi ca sinh vin', fontsize=16)
plt.xlabel('im s', fontsize=12)
plt.ylabel('S lng sinh vin', fontsize=12)

# Hin th biu 
plt.show()
```
